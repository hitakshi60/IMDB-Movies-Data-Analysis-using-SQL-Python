Director Analysis

```python
cur.execute("Select name,count(*) AS 'Total Movies',SUM(revenue) AS 'Total Revenue' FROM directors join movies ON movies.director_id=directors.id GROUP BY director_id ORDER BY SUM(revenue) desc")
director_analysis=cur.fetchall()
director_analysis1 = pd.DataFrame(director_analysis,columns=['name','Total Movies','Total Revenue'])
director_analysis1
```

```python
cur.execute("Select name,COUNT(title) AS 'Total Movies',SUM(revenue) AS 'Total Revenue' FROM directors join movies ON movies.director_id=directors.id GROUP BY director_id ORDER BY COUNT(title) desc")
director_analysis=cur.fetchall()
director_analysis1 = pd.DataFrame(director_analysis,columns=['name','Total Movies','Total Revenue'])
director_analysis1
```
